The next function is used to calculate the metrics given in ChartOCR paper. Make sure to verify the accuracy of the paths. Here, "gt" refers to the ground truth annotations and "pd" refers to the predicted json outputs.

```python
import json

# Function to measure the difference b/w the predictions and ground truth
def dist_p(pred_i , gt_i):

  if not gt_i and not pred_i:
    return 1
  elif not gt_i and pred_i:
    return 0
  elif gt_i and not pred_i:
    return 0
  else:
    for i,bbox_gt in enumerate(gt_i):
      try:
        bbox_pred = pred_i[i]

        dm = min(1, (abs(bbox_pred[0] - bbox_gt[0])/bbox_gt[2]) + abs((bbox_pred[1] - bbox_gt[1])/bbox_gt[3]) + abs((bbox_pred[3] - bbox_gt[3])/bbox_gt[3]) )
        return 1- (dm/len(gt_i))
      except:
        return 0


gt = json.load(open( ROOT_DIR + '/bardata/bar/annotations/instancesBar(1031)_test2019.json'))
pd = json.load(open( DEEPRULE +'/save/excel/bar.json'))

print(len(pd))
gt_dict = {}

net_f1 = 0
cnt = 0

for i in gt['images']:
  id = i['id']

  gt_bbox = []
  for j in gt['annotations']:
    if j['image_id'] == id:
      bbox_i = j['bbox']
      gt_bbox.append(bbox_i)


  f1_i = dist_p(pd[i['file_name']], gt_bbox)
  print(f1_i)

  if f1_i!=None:
    net_f1 += f1_i
    cnt += 1
 

score = net_f1/cnt
print("Final Average F1 for Excel Bar class using Pretrained Model :",score)
print("**************************************************************************************************************************************")
```

### Pretrained model: Testing Line class of ExcelChart400K dataset

NOTE: Metric for the line class has been inspired from the ChartInfo metric.

```python
!cd $DEEPRULE && python $DEEPRULE/test_pipe_type_cloud.py --image_path=$ROOT_DIR/linedata/line/images/test2019 --data_dir=$ROOT_DIR/linedata --base_dir=$ROOT_DIR  --cache_path=$ROOT_DIR/linedata --save_path=$DEEPRULE/save/excel/line.json
```

The next function is used to calculate the metrics given in ChartOCR paper. Make sure to verify the accuracy of the paths. Here, "gt" refers to the ground truth annotations and "pd" refers to the predicted json outputs.

```python
# Note: This evaluation function is inspired from ChartInfo competition Metric-6a.

import os
import sys
import json
import math
import itertools
import editdistance
import numpy as np
import scipy.optimize
import scipy.spatial.distance



def conv_gt(gt):
  out = {}
  for ind in gt['images']:
    i = ind['id']
    lines = []
    for j in gt['annotations']:
      if j['image_id'] == i:
        line = []
        point = None
        for ind_k,k in enumerate(j['bbox']):

          if ind_k%2 == 0:
            point = {}
            point["x"] = k
          elif ind_k%2 == 1:
            point["y"] = k
            line.append(point)
            

        lines.append(line)
        

    out[ind['file_name']] = lines
  
  return out


def conv_pd(pd):
  out = {}
  for ind in pd:
    i = pd[ind]
    lines = []
    for j in i:
      line = []
      for k in j:
        try:
          point = {"x": k[0], "y": k[1]}
          line.append(point)
        except:
          continue

      lines.append(line)
    out[ind] = lines
  return out


def check_groups(ds):
    try:
        _i = ds[0][0]
        return 1
    except Exception:
        return 0

def get_cont_recall(p_xs, p_ys, g_xs, g_ys, epsilon):
    total_score = 0
    total_interval = 0

    for i in range(g_xs.shape[0]):
        x = g_xs[i]
        
        if g_xs.shape[0] == 1:
            interval = 1
        elif i == 0:
            interval = (g_xs[i+1] - x) / 2
        elif i == (g_xs.shape[0] - 1):
            interval = (x - g_xs[i-1]) / 2
        else:
            interval = (g_xs[i+1] - g_xs[i-1]) / 2

        y = g_ys[i]
        y_interp = np.interp(x, p_xs, p_ys)
        error = min(1, abs( (y - y_interp) / (abs(y) + epsilon)))
        total_score += (1 - error) * interval
        total_interval += interval

    if g_xs.shape[0] != 1:
        assert np.isclose(total_interval, g_xs[-1] - g_xs[0])
    return total_score / total_interval

def compare_continuous(pred_ds, gt_ds):
    pred_ds = sorted(pred_ds, key=lambda p: float(p['x']))
    gt_ds = sorted(gt_ds, key=lambda p: float(p['x']))

    if not pred_ds and not gt_ds:
        return 1.0
    elif not pred_ds and gt_ds:
        return 0.0
    elif pred_ds and not gt_ds:
        return 0.0

    p_xs = np.array([float(ds['x']) for ds in pred_ds])
    p_ys = np.array([float(ds['y']) for ds in pred_ds])
    g_xs = np.array([float(ds['x']) for ds in gt_ds])
    g_ys = np.array([float(ds['y']) for ds in gt_ds])

    epsilon = (g_ys.max() - g_ys.min()) / 100.
    recall = get_cont_recall(p_xs, p_ys, g_xs, g_ys, epsilon)
    precision = get_cont_recall(g_xs, g_ys, p_xs, p_ys, epsilon)

    return (2 * precision * recall) / (precision + recall) if (precision + recall) else 0.

def compare_line(pred_ds, gt_ds):
    is_grouped = check_groups(gt_ds)
    if is_grouped:
        score = np.zeros((len(gt_ds), len(pred_ds)))
        for iter_seq1 in range(len(gt_ds)):
            for iter_seq2 in range(len(pred_ds)):
                score[iter_seq1, iter_seq2] = compare_continuous(gt_ds[iter_seq1], pred_ds[iter_seq2])
        
        row_ind, col_ind = scipy.optimize.linear_sum_assignment(-score)
        score = score[row_ind, col_ind].sum()/len(gt_ds)
    else:
        print(gt_ds)
        score = compare_continuous(pred_ds, gt_ds)

    return score


gt = json.load(open( ROOT_DIR + '/linedata/line/annotations/instancesLine(1023)_test2019.json'))
pd = json.load(open( DEEPRULE +'/save/excel/line.json'))


pgt = conv_gt(gt)
pdn = conv_pd(pd)


cnt = 0
score = 0
for i in pgt:
  pd_i = pdn[i]
  gt_i = pgt[i]

  if pd_i == [] and gt_i == []:
    f1_i = 1
  elif pd_i == [] and gt_i != []:
    f1_i = 0
  elif pd_i != [] and gt_i == []:
    f1_i = 0
  else:
    try:
      f1_i = compare_line(pd_i,gt_i)
    except:
      f1_i = 0

  if f1_i == []:
    f1_i = 0

  print(f1_i)
  score += f1_i
  cnt += 1

print("Final Average F1 for Excel Line class using Pretrained Model :",score/cnt)

print("**************************************************************************************************************************************")
```

### Pretrained model: Testing Pie class of ExcelChart400K dataset

```python
!cd $DEEPRULE && python $DEEPRULE/test_pipe_type_cloud.py --image_path=$ROOT_DIR/piedata/pie/images/test2019 --data_dir=$ROOT_DIR/piedata --base_dir=$ROOT_DIR  --cache_path=$ROOT_DIR/piedata --save_path=$DEEPRULE/save/excel/pie.json
```

The next function is used to calculate the metrics given in ChartOCR paper. Make sure to verify the accuracy of the paths. Here, "gt" refers to the ground truth annotations and "pd" refers to the predicted json outputs.

```python
import numpy as np
np.set_printoptions(linewidth=np.inf)
import math


# Calculate euclidean distance b/w p1 and p2
def euclid(p1, p2):

    try:
      x1 = float(p1[0])
      y1 = float(p1[1])
    except:
      x1 = float(p1)
      y1 = float(p1)


    try:
      x2 = float(p2[0])
      y2 = float(p2[1])
    except:
      x2 = float(p2)
      y2 = float(p2)


    return math.sqrt((x1 - x2) ** 2 + (y1 - y2) ** 2)


def score_func(mat_s,i,j,pred_new, gt_new):

  try:
    px = pred_new[0]
    py = pred_new[1]
  except:
    px = pred_new
    py = pred_new

  try:
    gx = gt_new[0]
    gy = gt_new[1]
  except:
    gx = gt_new
    gy = gt_new

  score_ij_new = mat_s[i-1][j-1] + 1 - (euclid(pred_new, gt_new)/ math.sqrt(gx**2 + gy**2))
  new_score = max(mat_s[i-1][j],mat_s[i][j-1],score_ij_new)
  mat_s[i][j] = new_score


  return mat_s


def f1_pie(pred_i , gt_i):
  gt_i.reverse()
  

  if len(gt_i) == 0 and pred_i == None:
    return 1

  if len(gt_i) != 0 and pred_i == None:
    return 0

  if len(gt_i) != 0 and len(pred_i) == 0:
    return 0


  mat_s = np.zeros((len(pred_i)+1,len(gt_i)+1))

  score = 0 

  for i in range(min(len(pred_i),len(gt_i))):
    gt_p1 = [gt_i[i][0],gt_i[i][1]]
    gt_p2 = [gt_i[i][2],gt_i[i][3]]
    pred_p1 = pred_i[i][1]
    pred_p2 = pred_i[i][2]

    
    if i == min(len(pred_i)-1,len(gt_i)-1):
      mat_s = score_func(mat_s,i+1,i+1,pred_p1,gt_p1)
      final_s = mat_s[i+1][i+1]/(i+1)
      return final_s
    elif i == 0:
      mat_s = score_func(mat_s,i+1,i+1,pred_p1,gt_p1)
      mat_s = score_func(mat_s,i+2,i+1,pred_p1,gt_p1)
      mat_s = score_func(mat_s,i+1,i+2,pred_p1,gt_p1)
    else:
      mat_s = score_func(mat_s,i+1,i+1,pred_p1,gt_p1)
      mat_s = score_func(mat_s,i+2,i+1,pred_p1,gt_p1)
      mat_s = score_func(mat_s,i+1,i+2,pred_p1,gt_p1)
   
     

gt = json.load(open( ROOT_DIR + '/piedata/pie/annotations/instancesPie(1008)_test2019.json'))
pd = json.load(open( DEEPRULE +'/save/excel/pie.json'))

gt_dict = {}


cnt=0
f1_net = 0

for i in gt['images']:
  id = i['id']
  gt_bbox = []

  for j in gt['annotations']:
    if j['image_id'] == id:
      bbox_i = j['bbox']
      gt_bbox.append(bbox_i)
      
  
  f1_i = f1_pie(pd[i['file_name']],gt_bbox)

  if math.isinf(f1_i):
        break


  f1_net += f1_i
  cnt += 1
  print(f1_i)
  
  

print("Final Average F1 for Excel Pie class using Pretrained Model :",f1_net/cnt)
print("**************************************************************************************************************************************")
```

## Phase-3:

### The assumption/conventions in this project are very important. Please make sure that assumptions made by original code authors are kept, else the code will break pretty easily. We're using a different file for training and testing ensemble models.

1. To train an ensemble model, we'll train a new model on 5 disjoint sets sampled from UB-PMC training data.

2. We'll train each model sequentially. After each training completion, we'll clean up the working direcory of temporary files created during the training process. This is to make sure there's no conflict with the next training process. Although, that's not mandatory.

3. But we have to change the "train2019" data folder and annotation file to reflect the appropriate split data.

4. After the process, we'll end with 5 trained models.

5. Each training iterations is prefixed with the "split" number. This will change the naming pattern for those models. 

6. For example, First split (split-0) will have "split integer 0". Second split (split-1) will have "split integer 1" and subsequently.

7. Thus, model will be saved according to following pattern : "CornerNetLine_[SPLIT NUMBER]50000.pkl"

8. Thus, model from split-0 will be saved as "CornerNetLine_050000.pkl" and model from split-1 will be saved as "CornerNetLine_150000.pkl"

9. If models, iterations are different than "50000" than please change that to reflect something similar "CornerNetLine_X50000.pkl"

10. Similarly, annotations files are saved as "x_instancesLine(1023)_train2019.json", where x is split number. For example, split-0 will have "0_instancesLine(1023)_train2019.json". Thus, the name of the file needs to be changed to "instancesLine(1023)_train2019.json" only.

11. Image data for the split should be of the following pattern: "train2019" or "test2019"
